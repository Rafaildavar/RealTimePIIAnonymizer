# Тестирование модели на простом датасете из 100 строк

```python
import pandas as pd
import re
from masker import mask_pii, mask_pii_with_mapping
```

```python
from unmask import unmask_pii
```

```python
df_simple_100= pd.read_csv('data/simple_data_100.csv')
```

```python
df_simple_100.sample(10)
```

```python
df_simple_100['pred'] = None
df_simple_100['mapping'] = None

for idx, row in df_simple_100.iterrows():
    res = mask_pii_with_mapping(row['original'])
    df_simple_100.at[idx, 'pred'] = res.masked_text
    df_simple_100.at[idx, 'mapping'] = res.mapping
```

```python
df_simple_100[['original', 'pred', 'mapping']].sample(10)
```

```python
df_simple_100['unmask'] = df_simple_100.apply(lambda r: unmask_pii(r['pred'], r['mapping']),axis=1)
```

```python
df_simple_100[['original', 'pred', 'unmask']].sample(10)
```

```python
df_simple_100[df_simple_100.original!=df_simple_100.unmask]
```
